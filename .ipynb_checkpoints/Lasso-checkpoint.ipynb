```python
#functions to create differentiation matrices

def make_dif1(n):
    D = np.zeros([n-1, n])
    a = np.diag(np.ones(n-1))
#     print(a)
    D[:,:-1 ]-= a
    D[:, 1:]+=a
    return D
#     print(D)
def make_dif2(n):
    D = np.zeros([n-2,n])
    a = np.diag(np.ones(n-2))
    D[:, :-2] += a
    D[:, 2:] += a
    D[:, 1:-1] -= 2*a
    return D

def make_dif12(n):
    D1 = make_dif1(n)
    D2 = make_dif2(n)
    return np.vstack((D1,D2))

def make_dif1_sparse(n):
    x = np.tile([-1,1], n-1)
    i = np.repeat(np.arange(0, n-1),2)
    j = []
    for m in range(n-1):
        j.append(m)
        j.append(m+1)
        
    return copt.spmatrix(x, i, j)

def make_dif2_sparse(n):
    x = np.tile([1, -2, 1],n-2)
    i = np.repeat(np.arange(0, n-2), 3)
    j = []
    for m in range(n-2):
        j.append(m)
        j.append(m+1)
        j.append(m+2)
#     print(len(x), len(i), len(j))
    return copt.spmatrix(x,i,j)

def make_dif12_sparse(n):
    x1 = np.tile([-1,1], n-1)
    i1 = np.repeat(np.arange(0, n-1),2)
    x2 = np.tile([1, -2, 1],n-2)
    i2 = np.repeat(np.arange(n-1, 2*n-3), 3)
    x = np.hstack((x1, x2))
    i = np.hstack((i1, i2))
    j = []
    for m in range(n-1):
        j.append(m)
        j.append(m+1)
    for m in range(n-2):
        j.append(m)
        j.append(m+1)
        j.append(m+2)
#     print(len(x), len(i), len(j))
    return copt.spmatrix(x,i,j)
```

```python
copt.solvers.options['show_progress'] = False


def LCfilter(data, l):
    """
    An implementation of the LC filter.
    For the dual variable v, the optimization problem becomes:
    min 1/2 v.T D D.T v - y.T D.T v
    where D is the differentiation matrix (in this case of order 1)
    y is the data
    under the condition -\lambda \leq v \leq \lambda
    
    This function returns the denoised trend under this filter
    x* = y - D.T v
    
    roughly equivalent to Lasso
    """    
    n = len(data)
#     y = data
#     D = make_dif1(n)
    y = copt.matrix(data)
    D = make_dif1_sparse(n)
    
#     q = copt.matrix(-np.matmul(y.T, D.T).reshape(-1,1))
#     P = copt.matrix(np.matmul(D, D.T))
    q = (-y.T*D.T).T

    P = D*D.T
    
    In1 = np.identity(n-1)
    g = np.vstack((In1, -In1))
    G = copt.matrix(g)
    
    h1s = np.ones(n-1).reshape(-1,1)*l
    H = np.vstack((h1s, h1s))
    h = copt.matrix(H)
#     print(h.size)
    sol = copt.solvers.qp(P, q, G = G, h = h)
    
    
#     filtv = np.array(sol['x'])

#     print((D.T*sol['x']).size)
    return data - np.array(D.T*sol['x']).ravel()

def LTfilter(data, l):
    """
    Roughly equivalent to Ridge Regression
    """
    
    n = len(data)
    y = copt.matrix(data)
    D = make_dif2_sparse(n)
    
    q = (-y.T*D.T).T
    P = D*D.T
    
    In2 = np.identity(n-2)
    g = np.vstack((In2, -In2))
    G = copt.matrix(g)
    
    h1s = np.ones(n-2).reshape(-1,1)*l
    H = np.vstack((h1s, h1s))
    h = copt.matrix(H)
    
    sol = copt.solvers.qp(P, q, G = G, h = h)
#     print((D.T*sol['x']).size)
    return data-np.array(D.T*sol['x']).ravel()
#     filtv = np.array(sol['x'])
    
#     return data - np.matmul(D.T, filtv).ravel()

def LTCfilter(data, l1, l2):
    n = len(data)
    y = copt.matrix(data)
    D = make_dif12_sparse(n)
    #     y = data
#     D = make_dif12(n)
    
    Q = D*D.T
    R =-D*y
    #     Q = copt.matrix(np.matmul(D, D.T))
#     R = copt.matrix(-np.matmul(D, y).reshape(-1,1))
    
    h1 = l1*np.ones(n-1).reshape(-1,1)
    h2 = l2*np.ones(n-2).reshape(-1,1)
    H = np.vstack((h1,h1,h2,h2))
    h = copt.matrix(H)
    
    In1 = np.identity(n-1)
    In2 = np.identity(n-2)
    z0 = np.zeros((n-1, n-2))
    
    row1 = np.hstack((In1, z0))
    row2 = np.hstack((z0.T, In2))
    
    InLTC = np.vstack((row1, row2))
    g = np.vstack((InLTC, -InLTC))
    G = copt.matrix(g)
    

    
    sol = copt.solvers.qp(P = Q,q = R, G=G, h = h)
    
    filtv = np.array(sol['x'])
    return y - np.matmul(D.T, filtv).ravel()

def HPfilter(data, l):
    y = data
    n = len(y)
    D = make_dif2(n)
    I = np.identity(n)
    return np.matmul(np.linalg.inv(I + 2*l*np.matmul(D.T,D)),y)
```

We will try to implement the procedure in Bruder, Dao, and Roncalli, but sklearn exists so we will try to use that.

The assumption is that within each window $t\in[0, T]$ there is a time series such that $y_t = x_t + \epsilon_t$ where $x_t$ is the trend process and $\epsilon_t$ is the noies process. Under an $L_1$ filter the objective function becomes:$$\frac 1 2 \sum_{t=1}^n(y_t - x_t)^2 + \lambda \sum_{t=2}^{n-1}|x_{t-1}-2x_t + x_{t+1}|$$
Then we want to minimize this under an optimal $\lambda$ for a given time window of size T. Bruder et al find that $\lambda_{max} \propto T^{5/2}$, but since sklearn exisets it should be able to calibrate for $\lambda$ without necessarily using their procedure.

For some reason, unlike th epaper, we are unable to impose continuity even though we really should but then again all we care about is the trend.

A cross validation algorithm given is:  
For a given $T_1, T_2$:
- Divide historical data into m rolling test sets T_2^i (i = 1 ... m)
- For each test window T_2^i compute the statistic $\lambda_{max}^i$
- From all $\lambda_{max}^i$ compute the average and standard deviation.
- Compute boundaries $\lambda_1 = \bar{\lambda} - 2\sigma_\lambda$, $\lambda_2 = \bar{\lambda} + 2\sigma_\lambda$
- For j = 1 to j = n:
 - Let $\lambda_j = \lambda_1 (\frac{\lambda_2}{\lambda_1})^{j/n}$
 - Divide the historical data into p rolling tradining sets T_1^k (k = 1, ... p)
 - For k = 1 to k = p:
   - Run L_1 filter on each T_1^k
   - Forecast trend for adjacent test window T_2^k
   - Compute error for that $\lambda$ on the test window $e^k(\lambda_j)$
 - Compute total error $e(\lambda_j) = \sum_{k=1}^m e^k(\lambda_j)$
-Minimize total error to find optimal $\lambda^*$
-Run the $L_1$ filter with $\lambda = \lambda^*$

Since we are using lasso to denoise, I assume x is just a simple linear series of 0 to T.

We attempt to implement the global-local trend test as well as the backtest.  The wealth of a portfolio is given by:
$$W_{t+1} = W_t + W_t(\alpha_t^*(\frac{S_{t+1}}{S_t} - 1 ) + (1-\alpha_t^*)r_t) $$
with the additional condition that:
$$\alpha_t^* = max(min(\frac{1}{\lambda W_0} \frac{\mu_t}{\sigma_t^2}, \alpha_{min}), \alpha_{max})$$
The constraint given is $(\alpha_{min}, \alpha_{max}) = (-1,1)$. Though in practice this condition is always going to be $\alpha_T = -1$ or $\alpha_T = 1$. Switching the order of the max and the min constrains it within the bounds without setting it equal to


We define the volatility estimator for a given time to be:
$$ \hat{\sigma_t}^2 = \frac 1 T \sum_{i = t - T + 1}^t \log^2\frac{S_i}{S_i-1}$$

```python
pGSPC = GSPC['Adj Close']['1998-01-01':'2010-12-31']
bGSPC = np.log(pGSPC)


T2 = 130
T3 = 4*T2
T1 = 4*T2
```

## Differences between scikit-learn and L1 filtering:

The implementation of Lasso regression in sklearn is:
$$\min_w \frac 1 {2n_{samples}} \lvert\lvert Xw - y \rvert\rvert^2_2 + \alpha\lvert\lvert w \rvert\rvert_1$$

where $w = (w_1, ..., w_p)$ and any given $\hat{y} = w_0 + w_1x_1 + ... + w_px_p$

The implmentation of the L1 filter in Bruder et al. is:
$$\min \frac 1 2 \lvert\lvert y - x\rvert\rvert_2^2 + \lambda\lvert\lvert z \rvert\rvert_1$$
where $z = Dx$, where D is a discrete derivative matrix operator.  
We assume the time series is generated by a process such that $y_t = x_t + \epsilon_t$ of trend and noise.

It is quickly evident that the two are not equivalent. The sklearn Lasso includes both trend and noise within the penalty, and penalizes the value itself, not hte derivative.
Hence, we cannot use sklearn's Lasso regression as an L1 filter, and must use computationally intense convex optimization on every time point.

```python
lcWindow = T2
lcGSPC = LCfilter(bGSPC, 1e5)
lctrend = np.diff(lcGSPC, n=1)
xlc = np.arange(lcWindow, Tmax, 1)
W0 = 1
llc = 1
W = W0
amin = -1
amax = 1
# r = .02/252
r = 0
Wcs = []
exposc = []
trendc = []
for i in range(T1, len(bGSPC)-1):
    guess = LCfilter(bGSPC[i-T1:i], llocal)
    diff = np.diff(guess, n= 1)[-1]
    trendc.append(diff)
# for t in range(1, len(xma)-1):
#     daily = HPfilter
    varest = ( np.log(pGSPC[i-ltWindow+1:i].values / pGSPC[i-ltWindow:i-1].values)**2).sum()/lcWindow
    astar = diff/llc/W0/varest
    astar = min(max(astar, amin), amax)
    
    W = W + W*( astar*
            (pGSPC[i+1]/pGSPC[i] - 1)
            + 
            (1-  astar)*r
           )
    exposc.append(astar)
    Wcs.append(W)
print(W)
plt.plot(Wcs)
```
